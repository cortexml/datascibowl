```python
#Import libraries for doing image analysis
from pylearn2.utils.image_proc import random_transform
from skimage.io import imread
from sklearn.ensemble import RandomForestClassifier as RF
import glob
import os
from matplotlib import pyplot as plt
from matplotlib import colors
from pylab import cm
import numpy as np
import pandas as pd
from scipy import ndimage
from skimage.transform import resize
from skimage.feature import peak_local_max
# make graphics inline
%matplotlib inline
```

```python
from skimage import transform as skimtr
```

```python
# get the classnames from the directory structure
directory_names = list(set(glob.glob(os.path.join("../data/datascibowl","train", "*"))\
 ).difference(set(glob.glob(os.path.join("../data/datascibowl","train","*.*")))))
```

```python
example_file = glob.glob(os.path.join(directory_names[5],"*.jpg"))[1]
print example_file
im = imread(example_file, as_grey=True)
im = resize(im, (48, 48))
```

# ROTATION 

```python
im_rot = im.copy()
im_rot_resize = im.copy()
im_rot = skimtr.rotate(im_rot, 45, resize=False, cval=1.0) # if using cval=1, raw img vals maintained
im_rot_resize = skimtr.rotate(im_rot_resize, 45, resize=True, cval=im.mean()) #if using mean, centred image properties maintained

plt.imshow(im, cmap=cm.gray)
plt.show()
plt.imshow(im_rot, cmap=cm.gray)
plt.show()
plt.imshow(im_rot_resize, cmap=cm.gray)
plt.show()
```
